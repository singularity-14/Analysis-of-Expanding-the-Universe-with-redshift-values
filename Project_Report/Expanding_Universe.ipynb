# The Expanding Universe

# I have prepared the data from Wikipedia and formed a dataframe for the analysis of the expansion of the universe.

# I have Studied the Relationship between Redshift (z) Values and Distance of Astronomical Objects.

# In 1929, Edwin Hubble demonstrated that galaxy clusters and other astronomical objects are moving away from us. According to the Doppler effect, galaxies moving away from each other cause a stretching of the wavelengths of light emitted by these galaxies, shifting them toward the longer-wavelength, lower-energy end of the electromagnetic spectrum. This shift is observed as a displacement of spectral lines toward the red end and is known as cosmological redshift.

# Redshift values indicate that the incoming electromagnetic radiation from distant stars and galaxies reaches the Earth with their spectra shifted towards the red end of the spectrum. This shift occurs because as space expands, electromagnetic waves have to travel a greater distance to reach Earth.

# If galaxies were moving towards each other, then the incoming spectra would shift towards the blue end of the spectrum. However, Sir Edwin Hubble observed the opposite phenomenon.

# In summary, the larger the redshift, the greater the distance, and the faster the galaxy is moving away from us.

```python
#extracting the columns
a=df.loc[:,"Redshift"]
b=df.loc[:,"Light Years Distance"]
c=df.loc[:,"Name"]
```
