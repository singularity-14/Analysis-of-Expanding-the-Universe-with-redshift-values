# The Expanding Universe

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Different Astronomical Objects and Thier Redshift Values

```python
df=pd.read_excel(r"C:\Users\patel\Documents\Msc_AA_2nd_sem\1st_module\project\data\galaxy_clusters_final.xlsx")
```

```python
df
```

```python
#extracting the columns
a=df.loc[:,"Redshift\n"]
b=df.loc[:,"Light Years Distance"]
c=df.loc[:,"Name"]
```

```python
#converting columns into the array
redshift=np.array(a)
distance=np.array(b)
name=np.array(c)
redshift
distance
name
```

```python
#ploting the point graph of redshift values and thier distances
plt.figure(figsize=(8,8))
plt.scatter(redshift,distance,color="black")
plt.title("The Expanding Universe")
plt.xlabel("Redshift(z)")
plt.ylabel("Distance(light year)")
plt.legend(["Redshift(Expanding Rate)(z)"])
plt.savefig("Expanding universe_001")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(distance,redshift,color='blue',marker='.',markersize=15,markeredgecolor='red')
plt.title("Distant Astronomical objects",fontsize=20,color='red')
plt.xlabel("Distance(light year)")
plt.ylabel("Redshift(z)")
plt.legend(["Redshift(Expanding Rate)(z)"])
plt.savefig("Expanding universe_002")
plt.grid()
plt.show()
```

# Different Constellations and Thier Redshift values

```python
con=pd.read_excel(r"C:\Users\patel\Documents\Msc_AA_2nd_sem\1st_module\project\data\Constellation_redshift_values.xlsx")
```

```python
data_1=con.rename(columns={'z':'Redshift Value(z)','cluster':'Astronomical Clusters'})
data_1
```

```python
#dropping the nan value columns
data_2=data_1.drop([6,7],axis=0)
```

```python
#extracting the columns from dataframe
z=data_2.loc[:,"Astronomical Clusters"]
n=data_2.loc[:,"Redshift Value(z)"]
```

```python
#converting the data into array
cluster_name=np.array(z)
redshift_values=np.array(n)
```

```python
print(redshift_values)
print(cluster_name)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(cluster_name,redshift_values,linewidth=3,color='red',marker='*',markersize=10,markeredgecolor='black')
plt.title("Distant Atronomical Clusters",fontsize=20,color='green')
plt.xlabel("cluster_name")
plt.ylabel("redshift_values(z)")
plt.legend(["Redshift(z)"])
plt.savefig("expanding universe_004")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.bar(cluster_name,redshift_values,color='red')
plt.title("Distant Atronomical Clusters",fontsize=20,color='green')
plt.xlabel("cluster_name")
plt.ylabel("redshift_values")
plt.legend(["Redshift(z)"])
plt.savefig("Expanding universe_005")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(cluster_name,redshift_values,color='b')
plt.title("Distant Astronomical Clusters",fontsize=20,color='red')
plt.xlabel("cluster_name")
plt.ylabel("redshift_values")
plt.legend(["Redshift value(z)"])
plt.savefig("Expanding universe_006")
plt.grid()
plt.show()
```
